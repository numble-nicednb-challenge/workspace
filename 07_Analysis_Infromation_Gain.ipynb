# 2. 사용할 변수 추출

```python
use_vars = ['당기순이익증가율', '영업이익증가율', '총자산영업이익률', '총자산순이익률', '자기자본영업이익률',
            '순운전자본회전율', '타인자본회전율', '유보율', '자기자본비율', '당좌비율',
            '유동비율']

raw_data = pd.read_csv('./custom_data/all_finance_data.csv', encoding='cp949')
raw_data.columns
```

```python
use_data = raw_data.iloc[:, 0:2]
use_data
```

```python
for var in use_vars:
    for i in raw_data.columns:
        if var in i[:len(var)]:
            use_data[i] = raw_data[i]

print(len(use_data.columns))
print(use_data.columns)
```

# 3. 내보내기

```python
use_data.to_csv('./custom_data/use_finance_data.csv', encoding='cp949')
```
