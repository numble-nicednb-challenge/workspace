# 1. 피처엔지니어링 3 : 지역특화산업 여부 파악하기

1990년말~2000년 초반부터 우리나라에서는 지역 균형 발전을 명분으로 수도권(서울, 경기, 인천)을 제외한 지방 기업을 대상으로 지역특화산업 육성사업을 별도의 예산으로 지원하고 있으며 창업기업 및 스타트업들이 많은 지원을 받을 수 있는 제도이다.

특정 지역의 지역특화산업에 해당하는 기업이라면, 혜택을 받을 가능성이 높아지고 이는 휴폐업의 여부와 연관이 있을 것이라고 예상한다. 따라서 우리는 연도별 지역특화산업을 조사하고, 해당하는 기업과 해당하지 않는 기업을 나누는 파생변수를 생성하려 한다.

# 2. 데이터 수집
지역특화산업은 KIAT에서 관리하다가 2022년부터 중소벤처기업부에서 관리한다. 우리가 필요한 데이터는 2019~2021년의 데이터이므로 KIAT 홈페이지에서 연도별 각 지역의 지역특화산업 공고문을 다운받아 수기로 옮기는 작업을 통해 지역특화산업의 핵심산업코드를 얻을 수 있었다.

# 3. 지역특화산업의 타당성 확인

지역특화산업이라고 무조건 폐업에 유리하진 않을 것이다. 따라서 우리는 연도별 지역특화산업 여부에 대한 파생변수를 이진 명목형 변수로 만들고, 동일하게 이진 명목형 변수인 휴폐업 구분과 연관성이 있는지 알아보기 위해 Two Sample Chi-Test를 진행해본다.

## 3-1. 라이브러리 불러오기

```python
import pandas as pd
from scipy import stats
import json
import warnings

warnings.filterwarnings('ignore')
```

## 3-2. 데이터 불러오기

```python
df = pd.read_csv('./custom_data/finance_address_indurstry.csv')
```

```python
df.shape
```

## 3-3. 카이제곱검정을 실시하기 위한 형태로 데이터 변환

```python
def change_target(value):
    """
    타겟값을 변환하기 위한 함수입니다.
    --------------------------------
    input = 휴폐업구분
    return = 폐업 or 휴업
    --------------------------------
    """
    if value in ['폐업', '휴업']:
        return '폐업'
    return '영업'

df['휴폐업구분'] = df['휴폐업구분'].apply(change_target)
```

```python
def extract_year(value):
    """
    결산년월에서 년도만 추출하는 함수입니다.
    -------------------------------------
    input = 결산년월
    return = 년도
    """
    return str(value)[:4]
df['년도'] = df['결산년월'].apply(extract_year)
```

```python
# 산업코드를 정수로 변환
df['산업코드1'] = df['산업코드1'].astype(int)
```

```python
# 필요한 비재무 피처만 추출
df_split = df[['사업자등록번호', '년도', '시도', '시군구', '산업코드1', '업종중분류', '업종대분류', '휴폐업구분']]
```

```python
df_split.head()
```

```python
# 시군구 중 오타, 없어진 지역이 있어 올바른 값으로 변환

missing_address = {
    '천산이':'천안시',
    '부천':'부천시',
    '아산군':'아산시',
    '청원군':'청주시'
}

def change_sigungoo(value):
    """
    시군구 컬럼에 잘못된 값들이 있어 올바른 값으로 변환해주는 함수입니다.
    ---------------------------------------------------------------
    input = 시군구
    return = 올바른 값으로 변환된 시군구, 원래 올바른 시군구
    ---------------------------------------------------------------
    """
    if value in missing_address.keys():
        return missing_address[value]
    return value

df['시군구'] = df['시군구'].apply(change_sigungoo)
```

## 3-4. 문제점 1) 수도권의 특화산업
- 지역특화산업 육성사업의 시작 계기가 수도권보다 비수도권의 발전이 더뎌 시행된 것이므로 수도권은 특화산업이 없다.
- 따라서 우리는 수도권에 가장 많이 분포된 업종 중 n개를 특화산업으로 지정하여 해당 여부를 따진다.
- 수도권은 비수도권보다 많은 발전을 한 것으로 보이므로 비수도권의 특화산업 개수가 가장 적은 것과 동일한 개수로 n을 지정한다.
    - 만약 강원도의 특화산업이 19개로 가장 적었다면, 수도권은 업종의 분포를 내림차순 기준으로 19개를 추출하여 수도권의 특화산업으로 지정한다는 의미

```python
# 비수도권의 지역특화산업 산업코드 불러오기
with open ("./custom_data/2019_flagship_indurstry.json", "r", encoding='UTF8') as f:
    flagship_indurstry_code_2019 = json.load(f)

with open ("./custom_data/2020_flagship_indurstry.json", "r", encoding='UTF8') as f:
    flagship_indurstry_code_2020 = json.load(f)

with open ("./custom_data/2021_flagship_indurstry.json", "r", encoding='UTF8') as f:
    flagship_indurstry_code_2021 = json.load(f)
```

```python
capital_area_flagship_indurstry_code_2019 = {}
capital_area_flagship_indurstry_code_2020 = {}
capital_area_flagship_indurstry_code_2021 = {}

for year, flagship_indurstry_code, captial_area_dict in zip(['2019', '2020', '2021'], [flagship_indurstry_code_2019, flagship_indurstry_code_2020, flagship_indurstry_code_2021], 
                                                             [capital_area_flagship_indurstry_code_2019, capital_area_flagship_indurstry_code_2020, capital_area_flagship_indurstry_code_2021]):
    length = []
    for _, val in flagship_indurstry_code.items():
        length.append(len(val))
    min_indurstry_count = min(length)
    print(f"{year}년 비수도권의 지역특화산업 산업코드 최솟값: {min_indurstry_count}개")
    df_year = df.loc[df['년도'] == year]
    print(f"{year}년 서울특별시의 업종중분류 종류: {df_year[df['시도'] == '서울특별시']['업종중분류'].nunique()}개")
    captial_area_dict['서울특별시'] = df_year[df['시도'] == '서울특별시']['업종중분류'].value_counts()[:min_indurstry_count].index

    print(f"{year}년 인천광역시의 업종중분류 종류: {df_year[df['시도'] == '인천광역시']['업종중분류'].nunique()}개")
    captial_area_dict['인천광역시'] = df_year[df['시도'] == '인천광역시']['업종중분류'].value_counts()[:min_indurstry_count].index

    print(f"{year}년 경기도의 업종중분류 종류: {df_year[df['시도'] == '경기도']['업종중분류'].nunique()}개")
    captial_area_dict['경기도'] = df_year[df['시도'] == '경기도']['업종중분류'].value_counts()[:min_indurstry_count].index
    
    print("===================================================================")
```

```python
# 수도권과 비수도권 딕셔너리 합치기
flagship_indurstry_code_2019.update(capital_area_flagship_indurstry_code_2019)
flagship_indurstry_code_2020.update(capital_area_flagship_indurstry_code_2020)
flagship_indurstry_code_2021.update(capital_area_flagship_indurstry_code_2021)
```

```python
# 반복 작업을 위한 딕셔너리 생성
capital_and_non_capital_total_dict = {
    '2019':flagship_indurstry_code_2019,
    '2020':flagship_indurstry_code_2020,
    '2021':flagship_indurstry_code_2021
}
```

```python
def centeral_indurstry(df):
    """
    지역특화산업 파생변수를 생성하는 함수입니다.
    ---------------------------------------------------------
    input = 년도, 시도, 업종중분류, 산업코드1이 있는 데이터프레임
    return = 지역특화산업에 해당하는 경우 1, 아닌 경우 0
    """
    year = df['년도']
    sido = df['시도']
    if sido in ['서울특별시', '인천광역시', '경기도']:
        if df['업종중분류'] in capital_and_non_capital_total_dict[year][sido]:
            return 1
        return 0
    else:
        if df['산업코드1'] in capital_and_non_capital_total_dict[year][sido]:
            return 1
        else:
            return 0

# 지역특화산업 파생변수 생성           
df['지역특화산업'] = df.apply(lambda x: centeral_indurstry(x), axis=1)
```

## 3-5. 년도별 지역특화산업 여부와 휴폐업 여부의 연관성 파악하기: Two Sample Chi-Test

```python
# 년도별로 파악하기 위한 데이터프레임 나누기

df_2019 = df.loc[df['년도'] == '2019']
df_2020 = df.loc[df['년도'] == '2020']
df_2021 = df.loc[df['년도'] == '2021']
```

```python
# 검정을 위한 교차표 생성

data_2019 = pd.crosstab(df_2019['지역특화산업'], df_2019['휴폐업구분'])
data_2020 = pd.crosstab(df_2020['지역특화산업'], df_2020['휴폐업구분'])
data_2021 = pd.crosstab(df_2021['지역특화산업'], df_2021['휴폐업구분'])
```

```python
def calculate_chi_test(data):
    """
    Two Sample Chi-Test를 수행하는 함수입니다.
    -----------------------------------------
    input = 지역특화산업 여부와 휴폐업 여부의 교차표
    return = 교차표에 카이제곱검정으로 산출한 p-value
    -----------------------------------------
    """
    _, p_value, _, _ = stats.chi2_contingency(data, correction=False)
    return p_value

for idx, data in enumerate([data_2019, data_2020, data_2021]):
    p_value = calculate_chi_test(data)
    if p_value > 0.05:
        print(f"{idx+2019}년의 지역특화산업 여부는 p-value가 {round(p_value, 3)}으로 0.05보다 크기 때문에 휴폐업 여부와 연관이 없습니다.")
    else:
        print(f"{idx+2019}년의 지역특화산업 여부는 p-value가 {round(p_value, 3)}으로 0.05보다 작기 때문에 휴폐업 여부와 연관이 있습니다.")
```

```python
df_2019 = df_2019[['사업자등록번호', '결산년월', '지역특화산업']]
df_2020 = df_2020[['사업자등록번호', '결산년월', '지역특화산업']]
df_2021 = df_2021[['사업자등록번호', '결산년월', '지역특화산업']]
```

## 3-5. 전체 데이터에서 지역특화산업 여부와 휴폐업 여부의 연관성 파악하기: Two Sample Chi-Test

```python
data_total = pd.crosstab(df['지역특화산업'], df['휴폐업구분'])
p_value_total = calculate_chi_test(data_total)

if p_value_total > 0.05:
        print(f"전체 데이터에서 지역특화산업 여부는 p-value가 {round(p_value_total, 3)}으로 0.05보다 크기 때문에 휴폐업 여부와 연관이 없습니다.")
else:
    print(f"전체 데이터에서 지역특화산업 여부는 p-value가 {round(p_value_total, 3)}으로 0.05보다 작기 때문에 휴폐업 여부와 연관이 있습니다.")
```
