# 4. 통계검정

```python
def change_target(value):
    """
    타겟값을 변환하기 위한 함수입니다.
    --------------------------------
    input = 휴폐업구분
    return = 폐업 or 휴업
    --------------------------------
    """
    if value in ['폐업', '휴업']:
        return '폐업'
    return '영업'

industrial_accident_yn['휴폐업구분'] = industrial_accident_yn['휴폐업구분'].apply(change_target)
```

```python
def change_yn(value):
    """
    타겟값을 변환하기 위한 함수입니다.
    --------------------------------
    input = 산업재해여부
    return = 해당 or 미해당
    --------------------------------
    """
    if value == 1:
        return '해당'
    return '미해당'

industrial_accident_yn['산업재해'] = industrial_accident_yn['산업재해'].apply(change_yn)
```

```python
# 년도별로 파악하기 위한 데이터프레임 나누기

df_2019 = industrial_accident_yn.loc[industrial_accident_yn['년도'] == '2019']
df_2020 = industrial_accident_yn.loc[industrial_accident_yn['년도'] == '2020']
df_2021 = industrial_accident_yn.loc[industrial_accident_yn['년도'] == '2021']
```

```python
# 검정을 위한 교차표 생성

data_2019 = pd.crosstab(df_2019['산업재해'], df_2019['휴폐업구분'])
data_2020 = pd.crosstab(df_2020['산업재해'], df_2020['휴폐업구분'])
data_2021 = pd.crosstab(df_2021['산업재해'], df_2021['휴폐업구분'])
```

```python
from scipy import stats


def calculate_chi_test(data):
    """
    Two Sample Chi-Test를 수행하는 함수입니다.
    -----------------------------------------
    input = 산업재해해당 여부와 휴폐업 여부의 교차표
    return = 교차표에 카이제곱검정으로 산출한 p-value
    -----------------------------------------
    """
    _, p_value, _, _ = stats.chi2_contingency(data, correction=False)
    return p_value

for idx, data in enumerate([data_2019, data_2020, data_2021]):
    p_value = calculate_chi_test(data)
    if p_value > 0.05:
        print(f"{idx+2019}년의 산업재해해당 여부는 p-value가 {round(p_value, 3)}으로 0.05보다 크기 때문에 휴폐업 여부와 연관이 없습니다.")
    else:
        print(f"{idx+2019}년의 산업재해해당 여부는 p-value가 {round(p_value, 3)}으로 0.05보다 작기 때문에 휴폐업 여부와 연관이 있습니다.")
```

```python
data_total = pd.crosstab(industrial_accident_yn['산업재해'], industrial_accident_yn['휴폐업구분'])
p_value_total = calculate_chi_test(data_total)

if p_value_total > 0.05:
        print(f"전체 데이터에서 산업재해 여부는 p-value가 {round(p_value_total, 3)}으로 0.05보다 크기 때문에 휴폐업 여부와 연관이 없습니다.")
else:
    print(f"전체 데이터에서 산업재해 여부는 p-value가 {round(p_value_total, 3)}으로 0.05보다 작기 때문에 휴폐업 여부와 연관이 있습니다.")
```
